```python
# Importing necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Assign the url to a variable
url = 'https://www.jobs.cz/prace/praha/?q%5B%5D=data%20analyst&locality%5Bradius%5D=0'
```

```python
# Sending request to Jobs.cz
page = requests.get(url, verify=False)
```

```python
# Parsing HTML page
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
# Finding all HTML code containing Data Analyst vacancies and assigning it to a variable

vac_info = soup.find_all('article', class_='SearchResultCard')

# Creating an empty list where all the scraped information will be inserted

vacancies = []

# Creating a for loop to repeate the same steps for every job insertion
    
for vacancy in vac_info:
    job_title = vacancy.header.h2.text.strip().replace(' - ',',')
    job_link = vacancy.header.h2.a['href']
    company = vacancy.find_all('li')[0].text.strip().replace(',','')       
    location = vacancy.find_all('li', class_='SearchResultCard__footerItem')[1].text.strip().replace(' – ',', ')
    
     # Appending all the scraped information to the vacancies list 
    
    vacancies.append((job_title, job_link, company, location))
    
    # Inserting all the information contained in the list into a pandas data frame
    
    df = pd.DataFrame(vacancies, columns= ['Job_title','Job_link','Company','Location'])
    
print(df)
```

```python
# Exporting data frame to an excel file
df.to_excel(r'C:\Users\dorep\Downloads\Data_Analyst_vacancies_Prague.xlsx', index=False)
```
