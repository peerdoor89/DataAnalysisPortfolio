```python
# Importing necessary libraries
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Assign the url to a variable
url = 'https://www.nba.com/draft/2023/draft-board'
```

```python
# Sending request to Nba.com
page = requests.get(url, verify=False)
```

```python
# Parsing HTML page
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup.prettify())
```

```python
# Assigning the table to be scraped to a variable
table = soup.find_all('table', class_="Crom_table__p1iZz")[0]
print(table)
```

```python
# Assigning HTML code related to the table's columns to a variable
fields = soup.find_all('th')
print(fields)
```

```python
# Cleaning HTML code and inserting columns into a list
column_names = [name.text.strip() for name in fields]
print(column_names)
```

```python
# Finding all HTML code containing players' information

player_info = table.find_all('tr', class_="DraftBoardTable_row__dUsAS")     

# Creating an empty list where all the scraped information will be inserted

drafted_players = []

# Creating a for loop to repeate the same steps for every player

for player in player_info:
    pick_num = player.select('td')[0].text.strip()
    team_name = player.select('td')[1].text[:-3].strip()
    player_name = player.select('td')[2].a.text.strip()
    position = player.select('td')[3].text.strip()
    school_club = player.select('td')[4].text.strip()
    country = player.select('td')[5].text.strip()
    height = player.select('td')[6].text.strip()
    weight = player.select('td')[7].text[:-4].strip()
    
    # Appending all the scraped information to the drafted_players list 
    
    drafted_players.append((pick_num,team_name,player_name,position,school_club,country,height,weight))
    
    # Inserting all the information contained in the list into a pandas data frame
    
    df = pd.DataFrame(drafted_players, columns=column_names)

print(df)
```
