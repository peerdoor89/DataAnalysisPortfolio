```python
import pandas as pd
```

```python
import os
```

```python
path = r'C:\Users\dorep\Downloads\World_Happiness'
extension = 'csv'
```

```python
# Looking for csv files at the path indicated above
files = [file for file in os.listdir(path) if file.endswith(extension)]
```

```python
dfs = []
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    dfs.append(df)
```

```python
# Concatenating all World Happiness Report's files
df = pd.concat(dfs, ignore_index=True)
```

```python
# Filtering df and selecting first and last year in the dataset
WHR_2015_2023 = df[df['year'].isin([2015,2023])].sort_values(by= 'country', ascending=True).reset_index(drop=True)
```

```python
# Creating a new column which shows the percentage change in the last 8 years for every country
WHR_2015_2023['Happiness_Pct_change_2015_2023'] = WHR_2015_2023.groupby(['country'])['happiness_score'].pct_change()*100
```

```python
# Dropping null values and sorting from the highest to the lowest percent change value
Happiness_Pct_change = WHR_2015_2023[['country','Happiness_Pct_change_2015_2023']].dropna().sort_values(by='Happiness_Pct_change_2015_2023',ascending =False).reset_index(drop=True)
```

```python
top_20_country_pct_change = Happiness_Pct_change.head(20)
```

```python
# Plotting Top 20 Countries with highest increase in happiness between 2015 and 2023
plt.figure(figsize=(22,12))
sns.set_style(rc = {'axes.facecolor': 'lightsteelblue'})
ax = sns.barplot(data=top_20_country_pct_change, x='country', y='Happiness_Pct_change_2015_2023', palette='YlOrRd_d')
ax.set(title='Top 20 Countries with highest increase in happiness between 2015 and 2023')
ax.set_xlabel('Country')
ax.set_ylabel('Percent change Happiness')
plt.show()
```

```python
bottom_20_country_pct_change = Happiness_Pct_change.tail(20).reset_index(drop=True)
bottom_20_country_pct_change
```

```python
# Plotting Top 20 Countries with highest decrease in happiness between 2015 and 2023
plt.figure(figsize=(25,12))
ax = sns.barplot(data=bottom_20_country_pct_change, x='country', y='Happiness_Pct_change_2015_2023',palette='Greys_d')
ax.set(title='Top 20 Countries with highest decrease in happiness between 2015 and 2023')
ax.set_xlabel('Country')
ax.set_ylabel('Percent change Happiness')
plt.show()
```
